    - Convert „casts“ data to a graph
    - Create a node for each actor
    - Create an edge if two actors appeared in the same movie
    - Perform Social Network Analysis
    
    - Compute general statistics
        -- e.g. number of nodes and edges, density, number of components
    - Identify key players using centrality measures
    - Identify clusters/communities in graph
    - Compute „Kevin Bacon“ number for each actor with selected key player  e.g. top actors with the highest/lowest number average number
    - Visualise important aspects of the analysis
    - Describe top clusters/communities
    - Insert a file with the graph exported to a GEXF format
    - All computed values should be present as attributes

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from networkx.drawing.nx_agraph import graphviz_layout
import numpy as np
```

```python
file = open("./../data.csv", "r")
text = file.read()
file.close

text = text.replace('"', "")
print(text[:100])
file = open("./../data_new.csv", "w")
file.write(text)
file.close
```

### Data preparation

```python
cast = pd.read_csv("./../data_new.csv", names=["MovieCode","Movie","Actor", "Genre","RoleDescription","Other"], sep=";")
#display(cast.info())
cast["Other"] = cast["Other"].fillna("")
cast["Genre"] = cast["Genre"].fillna("")
cast["RoleDescription"] = [str(x).replace(",","").replace("R:","") for x in cast["RoleDescription"]]# + cast["Other"].replace(",","")
cast["RoleDescription"] = cast["RoleDescription"]+cast["Other"]
cast.drop(columns=["Other"], inplace=True)
display(cast.head())
```

```python
#print(list(cast["Genre"].unique())[:20])
#small_cast = cast[cast["Genre"] == "Inn"]
#print(list(small_cast["Movie"].unique())[:20])
#display(small_cast.head())

##data where movie or actor is NaN value
cast_nan = cast[cast.isna().any(axis=1)]
display(cast_nan.info())
##There is only a small number of these errors so I will drop them from the dataset
cast.drop(cast_nan.index, axis=0,inplace=True)
#droppin actor "s a"
a_s_ = cast[cast["Actor"] == 's a'].index
cast.drop(a_s_, axis=0,inplace=True)
display(cast.head())
```

### Graph creation

```python
G = nx.Graph()
print("Unique actors in dataset (graph nodes):",len(list(cast["Actor"].unique())))

actors = cast["Actor"].unique()
movies = cast["MovieCode"].unique()
G.add_nodes_from(actors) #adding nodes to graph G
```

```python
import itertools
edges = []
#picking combinations of actors (no repetitions) for each movie
for movie in movies:
    tups = []
    for k in itertools.combinations(list(cast[cast["MovieCode"]==movie]["Actor"]),2):
        if(k[0] != k[1]):
            tups.append(k)
    
    edges.append(tups)

for edgelist in edges:
    G.add_edges_from(edgelist) #adding edges to graf G
    
print("Count of added edges: " + str(G.number_of_edges()))
```

### Computing general statistics

```python
#dropping isolates
isolates = nx.isolates(G)
G.remove_nodes_from(list(nx.isolates(G)))
print("Count of nodes: " + str(G.number_of_nodes()))
print("Count of edges: " + str(G.number_of_edges()))
```

```python
#computing density: possible edges in undirected graph n(n-1)/2
graph_density = G.number_of_edges() / (G.number_of_nodes()*((G.number_of_nodes() - 1) / 2.))
print("Graph density: {:.3%}".format(graph_density))
#no reciprocity - we have undirected graph
##creating components
components = [c for c in nx.connected_components(G)]
print("Number of components in graph:", len(components))
## degree statistics
degree_stat = list(dict(G.degree()).values())
print("Degree mean:", np.mean(degree_stat))
print("Degree max:", np.max(degree_stat))
print("Degree min:", np.min(degree_stat)) #should be 1 as we dropped isolates
print("Degree median:", np.median(degree_stat))
print("Degree std:", np.std(degree_stat))
```

### Identifying key players by centrality measure


```python
degree_centrality_n = nx.degree_centrality(G)
eigenvector_centrality_n = nx.eigenvector_centrality(G)
#betweenness_centrality_n = nx.betweenness_centrality(G)
#closeness_centrality_n = nx.closeness_centrality(G)
```

```python
from itertools import islice
def getCentrality(name, N, nodes):
    nodes = list(nodes.items())
    nodes.sort(key=lambda tup: tup[1], reverse=True)
    sorted_names = [x[0] for x in nodes]
    top_N = sorted_names[:N+1]
    print("Top",N,"actors by",name,"centrality:")
    for a in top_N:
        print("\t",a)
    return top_N
    
```

```python
getCentrality('degree', 10, degree_centrality_n)
#getCentrality('betweenness', 10, betweenness_centrality_n)
key_players = getCentrality('eigenvector', 10, eigenvector_centrality_n)
```

```python
#number of cliques (only some sizes -  some are visible from graph)
sizes = [3,4,5,6,7]
cliqs = []
for s in sizes:
    cliques = nx.algorithms.community.k_clique_communities(G,s)
    cliques = [c for c in cliques]
    cliqs.append(cliques)
    print("Number of cliques of size",s,":",len(cliques))
```

```python
attributes = {}
for i, node in enumerate(G.nodes()):
    attributes[node] ={}
## adding clique Id to each node in clique of size 5
c_index = 0
for clique in cliqs[2]:
    for actor in clique:
        attributes[actor]["Id"] = actor
        attributes[actor]["Cli"] = c_index
    c_index += 1
```

```python
#print(attributes)
```

```python
def getKevinBacon(graph, myKB):
    for i,actor in enumerate(graph.nodes()):
        kb = np.inf
        try:
            distance = len(nx.shortest_path(G, source=actor, target=myKB))
            if(distance < np.inf and actor != myKB):
                #print(distance)
                attributes[actor]["KB"] = distance
            elif(actor == myKB):
                attributes[actor]["KB"] = -np.inf
            else:
                attributes[actor]["KB"] = -1 # instead of infinity
        except nx.exception.NetworkXNoPath:
            attributes[actor]["KB"] = -1
            pass
        
```

```python
getKevinBacon(G, key_players[1])
print("Getting Kevin Bacon numbers for each actor and", key_players[1])
```

```python
## adding attributes to graph
nx.set_node_attributes(G, attributes)
## exporting graph
nx.write_gexf(G,"./../res/Goriginal.gexf")
```

```python
## Finding top cluster and other communities
clusters = nx.connected_component_subgraphs(G)
c_max = 0
c_sec = 0
max_cluster = None
max_second = None
i = 0
for cluster in clusters:
    #print(len(cluster.nodes()))
    if(i == 1):
        max_second = cluster
        c_sec = len(cluster.nodes())
        i = 2
    c_nodes = len(cluster.nodes())
    if(i > 2 and c_nodes > c_sec):
        max_second = cluster
        c_sec = len(cluster.nodes())
    if(c_nodes > c_max):
        c_max = c_nodes
        max_cluster = cluster
    i+=1
    
    
        
print("Max cluster nodes:", len(max_cluster.nodes()))
print("2nd cluster nodes:", len(max_second.nodes()))
```

### Other smaller graphs for visualization purposes

```python
actorssmall = set()
Gsmall = nx.Graph()
for cli in nx.find_cliques(G):
    if len(cli) == 7:
        for a in cli:
            actorssmall.add(a)

Gsmall.add_nodes_from(list(actorssmall))
print("Actors in new graph", len(list(actorssmall)))
```

```python
import itertools
edges = []
#picking combinations of actors (no repetitions) for each movie
for movie in movies:
    tups = []
    possible_actors = list(cast[(cast["MovieCode"]==movie) & (cast["Actor"].isin(list(actorssmall)))]["Actor"])
    for k in itertools.combinations(possible_actors,2):
        if(k[0] != k[1]):
            tups.append(k)
    
    edges.append(tups)

for edgelist in edges:
    Gsmall.add_edges_from(edgelist) #adding edges to graf G
    
print("Count of added adges: " + str(Gsmall.number_of_edges()))
```

```python
plt.figure(figsize=(20, 15))
pos = graphviz_layout(Gsmall)
nx.draw(Gsmall,pos, with_labels=False, node_size=10)
plt.show()
```

```python
## exporting graph
nx.write_gexf(Gsmall,"Gsmall.gexf")
```
