```python
import numpy as np
import cv2
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
def load_image(path):
    im = cv2.imread("./imgs/"+path)
    return im

def save_image(img, path):
    cv2.imwrite("./results/"+path, img)
```

```python
lena = load_image("lena.jpg")
marlon = load_image("marlon.jpg")
peppers = load_image("peppers.png")
```

```python
def get_sep_vec(sigma,k):
    x,y = np.meshgrid(np.linspace(-1,1,k),np.linspace(-1,1,k))
    d = np.sqrt(x*x+y*y)
    mu = 0.0
    g = np.exp(-( (d-mu)**2 / ( 2.0 * sigma**2 ) ) )
    return g[k//2,:], g[k//2,:].T
```

```python
def get_kernel(sigma,k):
    x, y = np.meshgrid(np.linspace(-1,1,k), np.linspace(-1,1,k))
    d = np.sqrt(x*x+y*y)
    mu = 0.0
    g = np.exp(-( (d-mu)**2 / ( 2.0 * sigma**2 ) ) )
    return g
```

```python
v,vt = get_sep_vec(16,7);
ker = np.outer(v,vt)
print(ker.shape)
sns.heatmap(ker)
```

```python
g = get_kernel(16,7);
sns.heatmap(g)
```

```python
def convolution(img,k,sigma):
    n,m,c = img.shape
    final = np.zeros((n,m,c))
    
    # get kernel vectors (since gaussian is symmetrical they are identical)
    v,vt = get_sep_vec(sigma,k)
    #norm kernel vectors
    v /= sum(v)
    vt /= sum(vt)
    
    for channel in range(c):
        # adding padding to edges
        padded = np.pad(img[:,:,channel].copy(),k//2, 'reflect')
        d,e = padded.shape
        
        #empty array for new data
        first = np.zeros((d,e))
        
        #row vector
        for (x,y), pix in np.ndenumerate(first):
            if(x >= k//2 and x < n+k//2 and y >= k//2 and y < m+k//2):
                first[x,y] = int(sum(v*padded[x-k//2:1+x+k//2,y])) #vector 1 * original
            
        ## adding padding
        first[:k//2,:] = np.flip(first[k//2:k//2+k//2,:], axis=0)
        first[:,:k//2] = np.flip(first[:,k//2:k//2+k//2], axis=1)
        first[-(k//2):,:] = np.flip(first[d-2*(k//2):n+k//2,:], axis=0)
        first[:,(e-k//2):] = np.flip(first[:,e-2*(k//2):m+k//2], axis=1)
       
        
        

        
        #column vector
        second = np.zeros((d,e))
        for (x,y), pix in np.ndenumerate(second):
            if(x >= k//2 and x < n+k//2 and y >= k//2 and y < m+k//2):
                second[x,y] = int(sum(vt*first[x,y-k//2:1+y+k//2])) #vector 2 * covoluted data
        
        # cutting of padding
        final[:,:,channel] = second[(k//2):n+k//2,(k//2):m+k//2]
    
    a = first.astype(int)
    plt.imshow(a)
    plt.show()
    plt.imshow(padded)
    plt.show()
    
    final = final.astype(int)
    return final
```

```python
res = convolution(lena, 21, 1.2)
```

```python
images = [lena, marlon]
names = ["lena", "marlon"]
kernel_sizes = [3,9,21]
sigma = 1.6

for ker in kernel_sizes:
    for im, name in zip(images, names):
        res = convolution(im, ker, sigma)
        save_image(res,name+" "+str(ker)+".jpg")
```

```python
for ker in kernel_sizes:
    res = convolution(peppers, ker, sigma)
    save_image(res,"peppers "+str(ker)+".jpg")
```
